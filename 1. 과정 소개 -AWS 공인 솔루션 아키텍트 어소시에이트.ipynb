### AWS란?
- AWS(Amazon Web Services)는 클라우드 제공업체입니다
- 그들은 당신에게 당신이 사용할 수 있는 서버와 서비스를 제공합니다
- 수요와 확장이 용이합니다
- AWS는 시간이 지남에 따라 IT를 혁신적으로 변화시켰습니다
- AWS는 세계에서 가장 큰 웹 사이트 중 일부를 지원합니다
    - Amazon.com
    - 넷플릭스
    

- AWS는 스파게티와 같다. 모든 서비스가 상호연결되어 있기 때문
